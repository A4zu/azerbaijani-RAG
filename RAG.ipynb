```python

from langchain_community.document_loaders import PyPDFLoader

pdf_paths = ["/content/- Azərbaycan Respublikasının Cinayət Məcəlləsi.pdf", "/content/- Azərbaycan Respublikasının İnzibati Xətalar Məcəlləsi.pdf","/content/- Azərbaycan Respublikasının Əmək Məcəlləsi.pdf"]

docs = []
for path in pdf_paths:
    loader = PyPDFLoader(path)
    docs.extend(loader.load())
```

**Cleaning**

```python

clean_docs = []

for doc in docs:
    clean_text = doc.page_content.replace('\x00', 'i')
    clean_text = ' '.join(clean_text.split())

    # Only keep if there's meaningful content
    if len(clean_text.strip()) > 50:
        clean_docs.append(
            doc.__class__(page_content=clean_text, metadata=doc.metadata)
        )
```

```python
print(f"Orijinal: {len(docs)} səhifə")
print(f"Təmiz: {len(clean_docs)} səhifə")
print(f"Atılan: {len(docs) - len(clean_docs)} səhifə")

# Bir nümunəyə bax
print(clean_docs[10].page_content)
```

**Chunking**

```python

from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=300
)
splits = text_splitter.split_documents(clean_docs)
```

```python
for i in range(3, 7):
    print(f'Split {i}: {splits[i-1].page_content}\n')
```

**Retriever**

```python

from langchain_community.vectorstores import Chroma
from langchain_groq import ChatGroq
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
import transformers
transformers.logging.set_verbosity_error()
```

```python
embedding = HuggingFaceEmbeddings(
    model_name="intfloat/multilingual-e5-base",
    model_kwargs={'device': 'cuda'},
    encode_kwargs={'normalize_embeddings': True}
)

vectorstore = Chroma.from_documents(splits, embedding=embedding)

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 10}
)
```

**RAG Chain Set Up**


```python
from langchain_openai import ChatOpenAI
```

```python


from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_google_genai import ChatGoogleGenerativeAI
from operator import itemgetter
from google.colab import userdata
prompt_text = """
Sən Azərbaycan Respublikasının qanunvericiliyinə(Azərbaycan Respublikasının Cinayət Məcəlləsi, Azərbaycan Respublikasının İnzibati Xətalar Məcəlləsi,Azərbaycan Respublikasının Əmək Məcəlləsi ) əsaslanaraq
istifadəçinin hüquqi sualını cavablandıracaqsan.


Sənin vəzifən:
- İstifadəçinin hüquqi sualına yalnız verilmiş kontekstə əsasən cavab verməkdir.
- Cavab konkret və birbaşa sualla əlaqəli olmalıdır.
- Lazımsız ümumi məlumat, xəbərdarlıq və əlavə hüquqi nəzəriyyə yazma.

Qaydalar:
1. Cavab yalnız istifadəçinin verdiyi sualla bağlı olsun.
2. Sualla əlaqəli olan qanun, maddə və bəndləri göstər.
3. Hər maddəni qısa və sadə formada izah et.
4. Əgər istifadəçi hüquqlarını soruşursa:
   - hansı hüquqlara sahib olduğunu maddələrlə göstər
   - hansı addımları ata biləcəyini yaz
5. Əgər cərimə, müddət, il və ya məbləğ varsa:
   - rəqəmləri dəqiq göstər
6. Əgər kontekstdə cavab yoxdursa:
   - “Verilmiş kontekstdə bu suala dair məlumat tapılmadı” yaz.
7. Cavabı maksimum praktik və konkret saxla.
8. Cavabda yalnız kontekstdə olan məlumatlardan istifadə et.




Kontekst:
{context}

Sual:
{question}

Cavab (Azərbaycan dilində):
"""


prompt_template = PromptTemplate(
    input_variables=["context", "question"],
    template=prompt_text
)

API_KEY = userdata.get('OPENAI_API')
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    api_key=API_KEY
)



def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

def query_with_prefix(question):
    """Add 'query: ' prefix for E5 models"""
    return f"query: {question}"

rag_chain = (
    {"context": retriever | format_docs,
    "question": RunnablePassthrough()}
    | prompt_template
    | llm
    | StrOutputParser()
)
```

**Query Testing**

```python
from IPython.display import Markdown

def ask_question(question):
    response = rag_chain.invoke(f"query: {question}")
    return response

test_questions = [
    "İşdən haqsız yerə çıxarılsam hüquqlarım nələrdir?",
    "Oğurluq törətdikdə minimum cəza neçə ildir?",
    "Yol hərəkəti qaydalarını pozduqda cərimə nə qədərdir?",
]

print("Test sualları:")
for i, q in enumerate(test_questions, 1):
    answer = ask_question(q)
    display(Markdown(f"**{i}. Sual:** {q}\n\n**Cavab:** {answer}\n\n---"))
```
